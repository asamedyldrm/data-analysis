## Birlestirme (join) Islemleri

```python
import numpy as np
import pandas as pd
```

```python
a = np.random.randint(1,30, size=(5,3))
```

```python
df1 = pd.DataFrame(a, columns = ["var1","var2","var3"])
```

```python
df1
```

```python
df2 = df1 + 99
```

```python
df2
```

```python
pd.concat([df1, df2]) #index numaralarÄ± tekrar ediyor.
```

```python
pd.concat([df1,df2], ignore_index = True)
```

```python
df1.columns
```

```python
df2.columns = ["var1","var2","var3"]
```

```python
df2
```

```python
df1
```

```python
pd.concat([df1,df2], join= "inner", ignore_index = True)
```
